```python
import pandas as pd
import glob
import os

# Define the path to your .txt files
folder_path = 'SoundAQnet_scene_ISOPl_ISOEv_PAQ8DAQs_case_study'  # Replace with your actual path
files = glob.glob(os.path.join(folder_path, '*.txt'))

# Define column names
columns = ['pleasant', 'eventful', 'chaotic', 'vibrant', 'uneventful', 'calm', 'annoying', 'monotonous']
data = []

for file_path in files:
    group_id = os.path.splitext(os.path.basename(file_path))[0]
    group_id = group_id.split('_')[0]  # Extract the group ID from the filename
    with open(file_path, 'r') as f:
        lines = f.readlines()
        if len(lines) >= 3:
            last_row = lines[2].strip().split('\t')
            if len(last_row) == 8:
                data.append([group_id] + list(map(float, last_row)))

# Create DataFrame
df = pd.DataFrame(data, columns=['GroupID'] + columns)

# Example: Save to CSV
# df.to_csv('soundaqnet_case_study_results.csv', index=False)

df
```

```python
import pandas as pd
import glob
import os

# Pad naar map met .txt bestanden
folder_path = 'SoundAQ_PAQ_case_study_dense_hop'
files = glob.glob(os.path.join(folder_path, '*.txt'))

# Kolomnamen
columns = ['pleasant', 'eventful', 'chaotic', 'vibrant', 'uneventful', 'calm', 'annoying', 'monotonous']
data = []

for file_path in files:
    group_id = os.path.splitext(os.path.basename(file_path))[0].split('_')[0]
    
    try:
        with open(file_path, 'r') as f:
            first_line = f.readline().strip()
            values = first_line.split('\t')
            if len(values) == 8:
                float_values = list(map(float, values))
                data.append([group_id] + float_values)
            else:
                print(f"Skipping {file_path}: expected 8 values, got {len(values)}")
    except Exception as e:
        print(f"Error processing {file_path}: {e}")

# DataFrame aanmaken
df = pd.DataFrame(data, columns=['GroupID'] + columns)

# Eventueel opslaan
df.to_csv('dense_hop_combined_case_study_results.csv', index=False)

df
```
