```python
import pandas as pd
import os
import soundscapy as sspy
from soundscapy.databases import isd
from pathlib import Path
```

```python
ssid = sspy.isd.load()
ssid, excl = sspy.isd.validate(ssid, allow_paq_na=False)
ssid = ssid[ssid['Language'] != 'cmn']
ssid
```

```python
from soundscapy.surveys.survey_utils import rename_paqs, PAQ

source_dir = Path(r'C:\Users\pepij\OneDrive - Delft University of Technology\THESIS\data\10672568 (1)\all_wav')

ssid["audio_path"] = ssid["GroupID"].apply(lambda x: source_dir.joinpath(str(x) + ".wav"))
setbefore = set(ssid["GroupID"].values)
# Keep only rows where the file exists
ssid = ssid[ssid["audio_path"].apply(os.path.exists)]
setafter = set(ssid["GroupID"].values)
# Reset index after filtering
ssid.reset_index(drop=True, inplace=True)

dictionary = {}
for paq in PAQ:
    dictionary[paq.id] = paq.label

ssid = rename_paqs(ssid, dictionary)

ssid = ssid[['GroupID', 'pleasant', 'chaotic', 'vibrant', 'uneventful', 'calm', 'annoying', 'eventful', 'monotonous', 'audio_path']]

columns_to_convert = [
    "pleasant", "chaotic", "vibrant", "uneventful", 
    "calm", "annoying", "eventful", "monotonous"
]

ssid[columns_to_convert] = ssid[columns_to_convert].astype(float).values

# # Before training — scale your labels
# means = ssid[columns_to_convert].mean()
# stds = ssid[columns_to_convert].std()

# ssid[columns_to_convert] = (ssid[columns_to_convert] - means) / stds
```
