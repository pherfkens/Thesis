```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV files
train_metrics_path = "AD_CNN_decreased_conv_layers_monitor_pleasant_train_metrics.csv"
val_metrics_path = "AD_CNN_decreased_conv_layers_monitor_pleasant_val_metrics.csv"

train_df = pd.read_csv(train_metrics_path)
val_df = pd.read_csv(val_metrics_path)

# Display the first few rows of both dataframes to understand the structure
train_df.head()
```

```python

# Define correct PAQ columns
paq_mse_columns = [
    'val_pleasant_mse', 'val_eventful_mse', 'val_chaotic_mse',
    'val_vibrant_mse', 'val_uneventful_mse', 'val_calm_mse',
    'val_annoying_mse', 'val_monotonous_mse'
]

# Recalculate the mean PAQ MSE
val_df['val_pa_mean'] = val_df[paq_mse_columns].mean(axis=1)

# Plotting training loss and validation metrics
plt.figure(figsize=(14, 6))
# Round the epoch values to the nearest integer to represent each full epoch (1, 2, 3, ...)
train_df['int_epoch'] = train_df['epoch'].round().astype(int)

# Group by the integer epochs and calculate the mean of raw train loss for each group
mean_train_loss_by_epoch = train_df.groupby('int_epoch')['train_loss'].mean().reset_index()

# Plot total training loss
plt.subplot(1, 2, 1)
plt.plot(train_df['epoch'], train_df['train_loss'], label='Raw Train Loss', alpha=0.5)
plt.plot(mean_train_loss_by_epoch['int_epoch'], mean_train_loss_by_epoch['train_loss'], label='Mean Train Loss per Epoch', linewidth=2)
plt.xlabel('Epoch')
plt.ylabel('Train Loss')
plt.title('Raw vs Mean Training Loss (Per Epoch)')
plt.grid(True)
plt.legend()

# Plot specified validation metrics with correct PAQ mean
plt.subplot(1, 2, 2)
plt.plot(val_df['epoch'], val_df['val_event_auc'], label='Val Event AUC')
plt.plot(val_df['epoch'], val_df['val_pleasant_mse'], label='Val Pleasant MSE')
plt.plot(val_df['epoch'], val_df['val_pa_mean'], label='Mean 8D PAQ MSE')
plt.xlabel('Epoch')
plt.ylabel('Metric Value')
plt.title('Validation Metrics Over Epochs')
plt.grid(True)
plt.legend()

plt.tight_layout()

# Remove the known suffix
base_name = train_metrics_path.replace("_train_metrics.csv", ".pdf")
print(base_name)

# plt.savefig(f'{base_name}')  # Save as PDF
plt.show()
```
